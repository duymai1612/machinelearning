```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('dataset3.csv')
df.columns = df.columns.str.replace(' ', '')
df
```

```python
import math
median_fixedacidity=math.floor(df.fixedacidity.median())
median_volatileacidity=math.floor(df.volatileacidity.median())
median_citricacid=math.floor(df.citricacid.median())
median_residualsugar=math.floor(df.residualsugar.median())
median_chlorides=math.floor(df.chlorides.median())
median_pH=math.floor(df.pH.median())
median_sulphates=math.floor(df.sulphates.median())


median_fixedacidity
median_volatileacidity
median_citricacid
median_residualsugar
median_chlorides
median_pH
median_sulphates
```

```python
df.fixedacidity=df.fixedacidity.fillna(median_fixedacidity)
df.volatileacidity=df.volatileacidity.fillna(median_volatileacidity)
df.citricacid=df.citricacid.fillna(median_citricacid)
df.residualsugar=df.residualsugar.fillna(median_residualsugar)
df.chlorides=df.chlorides.fillna(median_chlorides)
df.pH=df.pH.fillna(median_pH)
df.sulphates=df.sulphates.fillna(median_sulphates)
df
```

```python
y = df[{'fixedacidity'}]
```

```python
x = df.drop(['type', 'fixedacidity'], axis =1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.3)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
print(lr.coef_)
```

```python
print(lr.intercept_)
```

```python
y_pred = lr.predict(x)
print('predicted response:', y_pred, sep='\n')
```

```python
from sklearn.metrics import mean_squared_error 
  
# # Given values 
# Y_true = [2,4,5,4,5]  # Y_true = Y (original values) 
  
# # calculated values 
# Y_pred = [2.8,3.4,4,4.6,5.2]  # Y_pred = Y' 
  
# Calculation of Mean Squared Error (MSE) 
mean_squared_error(y,y_pred) 
```

```python
RSS = np.sum((y-y_pred)**2)
print("Residual sum of squares: %.2f" % RSS)
```
